```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats

import requests
import time
from scrapy.selector import Selector
from scrapy.http import HtmlResponse
from selenium import webdriver
from bs4 import BeautifulSoup
import json

chromedriver = "/Users/amos/Desktop/chromedriver"
```

```python
# opens selenium webdriver
browser = webdriver.Chrome(executable_path= chromedriver)
```

```python
browser.get('http://stats.nba.com/league/player/#!/scoring/?sort=PCT_UAST_2PM&dir=1')
#json html format
HTML_logs = browser.page_source
years = Selector(text=HTML_logs).xpath("//select[@name='Season']/option/text()").extract()
years
```

```python
stats_url = 'http://stats.nba.com/stats/leaguedashplayerstats?College=&Conference=&Country=\
&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&GameScope=&GameSegment=&Height=&LastNGames=0\
&LeagueID=00&Location=&MeasureType=Scoring&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N\
&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=%s\
&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&StarterBench=&TeamID=0&VsConference=&VsDivision=&Weight='
```

```python
# browser = webdriver.Chrome(executable_path= chromedriver)
json_text = None
for year in years:
    print year
    link = stats_url % year
    #selenium browser opens link
    browser.get(link)
    HTML_logs = browser.page_source
    soup = BeautifulSoup(HTML_logs, 'lxml')
    get_text = soup.findAll('pre')[0].getText().encode('utf-8')
    json_text = json.loads(get_text)
   
    df = pd.DataFrame(json_text['resultSets'][0]['rowSet'], columns=json_text['resultSets'][0]['headers']) 
    save_csv = '/Users/amos/Desktop/DSI-SF-2-amoskim25/copy_capstone/datasets/unassisted_stats/unassisted_stats%s.csv' % year
    df.to_csv(save_csv)
    time.sleep(5*np.random.rand())
```
